## 1. Class và Object

- **Class** = bản thiết kế (blueprint)
- **Object** = sản phẩm tạo ra từ bản thiết kế

Ví dụ: Class `Car` là bản vẽ, mỗi chiếc xe thực tế là một Object.

```python
# Class đơn giản nhất
class Dog:
    pass

# Tạo object (instance)
d1 = Dog()
d2 = Dog()

print(f"d1: {d1}")
print(f"d2: {d2}")
print(f"Cùng kiểu: {type(d1) == type(d2)}")
print(f"Cùng object: {d1 is d2}")  # False - 2 object khác nhau
```

```python
# __init__ = Constructor (hàm khởi tạo)
# self = tham chiếu đến chính object hiện tại

class Student:
    def __init__(self, name, age, gpa):
        self.name = name    # Attribute (thuộc tính)
        self.age = age
        self.gpa = gpa

# Tạo object
s1 = Student("Hải", 20, 3.5)
s2 = Student("Lan", 21, 3.8)

print(f"s1: {s1.name}, {s1.age} tuổi, GPA={s1.gpa}")
print(f"s2: {s2.name}, {s2.age} tuổi, GPA={s2.gpa}")
```

### Class Attributes vs Instance Attributes

- **Instance Attributes:** Riêng biệt cho từng object (khai báo qua `self`).
- **Class Attributes:** Dùng chung cho tất cả các object của lớp đó (khai báo trực tiếp trong class).

```python
class Dog:
    species = "Canis familiaris"  # Class Attribute

    def __init__(self, name, age):
        self.name = name          # Instance Attribute
        self.age = age

buddy = Dog("Buddy", 9)
miles = Dog("Miles", 4)

print(f"{buddy.name} là loài {buddy.species}")
print(f"{miles.name} cũng là loài {miles.species}")

# Thay đổi Class Attribute sẽ ảnh hưởng đến tất cả
Dog.species = "Cảnh khuyển"
print(f"Sau khi đổi: {buddy.name} là {buddy.species}")
```

## 2. Methods (Phương thức)

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    
    def deposit(self, amount):
        """Nạp tiền"""
        if amount <= 0:
            print("⚠️ Số tiền phải > 0!")
            return
        self.balance += amount
        print(f"  +{amount:,}đ → Số dư: {self.balance:,}đ")
    
    def withdraw(self, amount):
        """Rút tiền"""
        if amount > self.balance:
            print("❌ Không đủ tiền!")
            return
        self.balance -= amount
        print(f"  -{amount:,}đ → Số dư: {self.balance:,}đ")
    
    def info(self):
        print(f"  Chủ TK: {self.owner}, Số dư: {self.balance:,}đ")

# Sử dụng
acc = BankAccount("Hải", 1000000)
acc.info()
acc.deposit(500000)
acc.withdraw(200000)
acc.withdraw(5000000)  # Không đủ tiền
```

```python
# ❌ LỖI: Quên self khi gọi attribute trong method
class Bad:
    def __init__(self, x):
        self.x = x
    
    def show(self):
        # print(x)    # ❌ NameError: x chưa định nghĩa
        print(self.x)  # ✅ Phải dùng self.x

b = Bad(42)
b.show()
```

## 3. Encapsulation (Đóng gói)

Kiểm soát truy cập vào attributes:

| Prefix | Ý nghĩa | Truy cập |
|--------|---------|----------|
| `name` | Public | Ở đâu cũng được |
| `_name` | Protected (quy ước) | Nên dùng trong class/subclass |
| `__name` | Private (name mangling) | Chỉ trong class |

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name          # Public
        self._department = "IT"    # Protected (quy ước)
        self.__salary = salary     # Private
    
    def get_salary(self):          # Getter
        return self.__salary
    
    def set_salary(self, value):   # Setter (có validation)
        if value < 0:
            print("❌ Lương không thể âm!")
            return
        self.__salary = value

emp = Employee("Hải", 10000000)

# Public - truy cập bình thường
print(f"Tên: {emp.name}")

# Protected - truy cập được nhưng KHÔNG NÊN
print(f"Phòng: {emp._department}")

# Private - không truy cập trực tiếp được
try:
    print(emp.__salary)
except AttributeError as e:
    print(f"❌ Private: {e}")

# ✅ Qua getter/setter
print(f"Lương (getter): {emp.get_salary():,}đ")
emp.set_salary(-100)  # Bị chặn!
emp.set_salary(15000000)
print(f"Lương mới: {emp.get_salary():,}đ")
```

```python
# ⚠️ Private trong Python chỉ là name mangling, KHÔNG thực sự private
# Python đổi __salary thành _Employee__salary
print(f"Hack: {emp._Employee__salary:,}đ")  # Vẫn truy cập được!
# → Python tin vào "responsible developers", không ép buộc
```

```python
# @property - cách Pythonic hơn cho getter/setter
class Circle:
    def __init__(self, radius):
        self._radius = radius
    
    @property
    def radius(self):
        return self._radius
    
    @radius.setter
    def radius(self, value):
        if value < 0:
            raise ValueError("Bán kính phải >= 0")
        self._radius = value
    
    @property
    def area(self):
        import math
        return math.pi * self._radius ** 2

c = Circle(5)
print(f"Bán kính: {c.radius}")   # Gọi như attribute, thực ra là getter
print(f"Diện tích: {c.area:.2f}")

c.radius = 10                     # Gọi setter
print(f"Bán kính mới: {c.radius}")

try:
    c.radius = -5  # Setter chặn!
except ValueError as e:
    print(f"❌ {e}")
```

## 4. Inheritance (Kế thừa)

Class con **kế thừa** attributes và methods từ class cha.

```python
# Class cha (Parent / Base)
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def speak(self):
        print(f"{self.name} makes a sound")
    
    def info(self):
        print(f"{self.name}, {self.age} tuổi")

# Class con (Child / Derived)
class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)  # Gọi __init__ của cha
        self.breed = breed           # Attribute riêng
    
    def speak(self):                 # Override method
        print(f"{self.name}: Gâu gâu! 🐕")

class Cat(Animal):
    def speak(self):                 # Override method
        print(f"{self.name}: Meo meo! 🐱")

# Sử dụng
d = Dog("Buddy", 3, "Golden")
c = Cat("Kitty", 2)

d.info()     # Kế thừa từ Animal
d.speak()    # Override
print(f"Giống: {d.breed}")  # Riêng của Dog
print()
c.info()     # Kế thừa từ Animal
c.speak()    # Override
```

```python
# isinstance() - kiểm tra kiểu
d = Dog("Rex", 5, "Husky")

print(f"isinstance(d, Dog): {isinstance(d, Dog)}")       # True
print(f"isinstance(d, Animal): {isinstance(d, Animal)}") # True (kế thừa!)
print(f"isinstance(d, Cat): {isinstance(d, Cat)}")       # False
```

```python
# ❌ LỖI: Quên gọi super().__init__()
class BadDog(Animal):
    def __init__(self, name, age, breed):
        # Quên super().__init__(name, age) !
        self.breed = breed

try:
    bd = BadDog("Rex", 5, "Husky")
    bd.info()  # AttributeError: 'BadDog' has no attribute 'name'
except AttributeError as e:
    print(f"LỖI: {e}")
    print("→ Phải gọi super().__init__() để khởi tạo attributes cha!")
```

### Đa kế thừa (Multiple Inheritance)

Một lớp con có thể kế thừa từ nhiều lớp cha cùng lúc.

```python
class Animal:
    def move(self):
        print("Đang di chuyển...")

class Mammal:
    def feed_milk(self):
        print("Cho con bú...")

class Bat(Animal, Mammal):
    def fly(self):
        print("Đang bay...")

batman = Bat()
batman.move()       # Từ Animal
batman.feed_milk()  # Từ Mammal
batman.fly()        # Của riêng Bat
```

## 5. Abstraction (Trừu tượng)

Sử dụng module `abc` để tạo lớp cơ sở trừu tượng. Lớp này không thể tạo object trực tiếp, mà bắt buộc các lớp con phải triển khai các phương thức trừu tượng.

```python
from abc import ABC, abstractmethod

class Computer(ABC):
    @abstractmethod
    def process(self):
        pass

class Laptop(Computer):
    def process(self):
        print("Laptop đang xử lý dữ liệu...")

class Desktop(Computer):
    def process(self):
        print("Desktop đang xử lý dữ liệu mạnh mẽ...")

devices = [Laptop(), Desktop()]
for d in devices:
    d.process()
```

## 6. Polymorphism (Đa hình)

Cùng tên method nhưng **hành vi khác nhau** tùy theo class.

```python
# Đa hình qua inheritance
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        import math
        return math.pi * self.r ** 2

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height

# Cùng gọi .area() nhưng kết quả khác nhau
shapes = [
    Rectangle(5, 3),
    Circle(4),
    Triangle(6, 4)
]

for shape in shapes:
    print(f"  {type(shape).__name__}: area = {shape.area():.2f}")
```

## 7. Phương Thức Đặc Biệt (Dunder Methods)

| Method | Khi nào gọi | Ví dụ |
|--------|------------|-------|
| `__init__` | Tạo object | `obj = Class()` |
| `__str__` | `print(obj)` hoặc `str(obj)` | In thân thiện |
| `__repr__` | `repr(obj)`, debug | In chi tiết |
| `__del__` | `del obj` | Khi object bị hủy |
| `__len__` | `len(obj)` | Độ dài |
| `__eq__` | `obj1 == obj2` | So sánh |
| `__lt__` | `obj1 < obj2` | Nhỏ hơn |
| `__add__` | `obj1 + obj2` | Cộng |

```python
# ⚠️ Không có __str__ → print xấu
class PointBad:
    def __init__(self, x, y):
        self.x = x
        self.y = y

p = PointBad(3, 4)
print(f"Không __str__: {p}")  # <__main__.PointBad object at 0x...>
```

```python
# ✅ Có __str__ → print đẹp
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Point({self.x}, {self.y})"
    
    def __repr__(self):
        return f"Point(x={self.x}, y={self.y})"
    
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def distance(self):
        return (self.x**2 + self.y**2) ** 0.5

p1 = Point(3, 4)
p2 = Point(1, 2)
p3 = Point(3, 4)

print(f"print: {p1}")              # __str__
print(f"repr: {repr(p1)}")         # __repr__
print(f"p1 == p3: {p1 == p3}")     # __eq__ → True
print(f"p1 == p2: {p1 == p2}")     # __eq__ → False
print(f"p1 + p2: {p1 + p2}")       # __add__
print(f"Khoảng cách: {p1.distance():.2f}")
```

## 8. Bài Tập Tổng Hợp

```python
# Bài: Quản lý danh sách sinh viên
class Student:
    def __init__(self, name, student_id, scores=None):
        self.name = name
        self.student_id = student_id
        self.scores = scores if scores else []
    
    def add_score(self, score):
        if 0 <= score <= 10:
            self.scores.append(score)
        else:
            print(f"⚠️ Điểm phải từ 0-10")
    
    def avg(self):
        if not self.scores:
            return 0
        return sum(self.scores) / len(self.scores)
    
    def grade(self):
        a = self.avg()
        if a >= 9: return "A"
        if a >= 8: return "B"
        if a >= 7: return "C"
        if a >= 5: return "D"
        return "F"
    
    def __str__(self):
        return f"{self.student_id} - {self.name} | TB={self.avg():.1f} ({self.grade()})"
    
    def __lt__(self, other):
        return self.avg() < other.avg()


class Classroom:
    def __init__(self, name):
        self.name = name
        self.students = []
    
    def add_student(self, student):
        self.students.append(student)
    
    def show_all(self):
        print(f"\n📋 Lớp {self.name} ({len(self.students)} SV):")
        for s in sorted(self.students, reverse=True):  # Dùng __lt__
            print(f"  {s}")
    
    def class_avg(self):
        if not self.students:
            return 0
        return sum(s.avg() for s in self.students) / len(self.students)


# Sử dụng
room = Classroom("PFP-K19")

s1 = Student("Hải", "SE001", [8.5, 9.0, 7.5])
s2 = Student("Lan", "SE002", [9.0, 9.5, 10.0])
s3 = Student("Minh", "SE003", [6.5, 7.0, 5.5])

room.add_student(s1)
room.add_student(s2)
room.add_student(s3)

room.show_all()
print(f"\nĐiểm TB lớp: {room.class_avg():.2f}")
```
