## 1. List (Danh sách)

- **Mutable** (có thể thay đổi)
- **Có thứ tự** (ordered)
- **Cho phép trùng lặp**
- Kí hiệu: `[ ]`

```python
# Tạo list
nums = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]
empty = []

print(f"nums: {nums}")
print(f"mixed: {mixed}")
```

```python
# Index và Slicing
fruits = ["apple", "banana", "cherry", "date"]
print(f"fruits[0]:  {fruits[0]}")
print(f"fruits[-1]: {fruits[-1]}")
print(f"fruits[1:3]: {fruits[1:3]}")
print(f"fruits[::-1]: {fruits[::-1]}")
```

```python
# Các phương thức quan trọng
nums = [3, 1, 4]
print(f"Ban đầu: {nums}")

nums.append(5)        # Thêm cuối
print(f"append(5): {nums}")

nums.insert(1, 9)     # Chèn tại index 1
print(f"insert(1,9): {nums}")

nums.remove(4)        # Xóa giá trị 4
print(f"remove(4): {nums}")

last = nums.pop()     # Xóa cuối, trả về giá trị
print(f"pop(): {last}, list: {nums}")
```

```python
# ⚠️ sort() vs sorted()
nums = [3, 1, 4, 1, 5]

# sort() → sửa tại chỗ, trả về None!
result = nums.sort()
print(f".sort(): result={result}, nums={nums}")

# sorted() → trả về list mới
nums = [3, 1, 4, 1, 5]
result = sorted(nums)
print(f"sorted(): result={result}, nums={nums}")
```

```python
# ⚠️ BẪY: Tham chiếu vs Sao chép
a = [1, 2, 3]
b = a           # b TRỎ ĐẾN cùng list!
b.append(4)
print(f"a = {a}")  # [1,2,3,4] ← a cũng bị đổi!

# ✅ Sao chép:
a = [1, 2, 3]
b = a.copy()    # Bản sao ĐỘC LẬP
b.append(4)
print(f"a = {a}")  # [1,2,3] ← không đổi ✓
```

```python
# List Comprehension
squares = [x**2 for x in range(1, 6)]
print(f"Bình phương: {squares}")

evens = [x for x in range(10) if x % 2 == 0]
print(f"Số chẵn: {evens}")
```

```python
# Hàm tích hợp
nums = [3, 1, 4, 1, 5, 9]
print(f"len: {len(nums)}, sum: {sum(nums)}, min: {min(nums)}, max: {max(nums)}")

# ❌ LỖI: Hàm trên list rỗng
try:
    min([])  # ValueError
except ValueError as e:
    print(f"LỖI min([]): {e}")
```

## 2. Tuple (Bộ)

- **Immutable** (KHÔNG thể thay đổi)
- **Có thứ tự** (ordered)
- Kí hiệu: `( )`
- Nhanh hơn list, an toàn hơn

```python
# Tạo tuple
t1 = (1, 2, 3)
t2 = ("An", 20, 8.5)   # Nhiều kiểu
t3 = (42,)              # ⚠️ Tuple 1 phần tử PHẢI có dấu phẩy
t4 = ()                 # Tuple rỗng

print(f"t1: {t1}, kiểu: {type(t1)}")
print(f"t3: {t3}, kiểu: {type(t3)}")

# ⚠️ Không có dấu phẩy → KHÔNG phải tuple!
not_tuple = (42)
print(f"\n(42) kiểu: {type(not_tuple)}")   # int!
print(f"(42,) kiểu: {type((42,))}")         # tuple ✓
```

```python
# Truy cập (giống list)
t = (10, 20, 30, 40, 50)
print(f"t[0]: {t[0]}")
print(f"t[-1]: {t[-1]}")
print(f"t[1:3]: {t[1:3]}")
```

```python
# ❌ LỖI: Tuple KHÔNG thể thay đổi!
t = (1, 2, 3)
try:
    t[0] = 99  # TypeError!
except TypeError as e:
    print(f"LỖI: {e}")

# Cũng không có append, remove, sort...
try:
    t.append(4)
except AttributeError as e:
    print(f"LỖI: {e}")
```

```python
# Tuple Packing / Unpacking

# Packing
point = 3, 4  # Không cần ngoặc
print(f"point: {point}, kiểu: {type(point)}")

# Unpacking
x, y = point
print(f"x={x}, y={y}")

# Ứng dụng: Swap 2 biến
a, b = 10, 20
a, b = b, a    # ← Tuple unpacking!
print(f"Sau swap: a={a}, b={b}")
```

```python
# ❌ LỖI: Unpacking sai số lượng
t = (1, 2, 3)
try:
    a, b = t  # ValueError: quá nhiều giá trị
except ValueError as e:
    print(f"LỖI: {e}")

# ✅ Dùng * để gom phần còn lại
a, *rest = t
print(f"a={a}, rest={rest}")
```

```python
# Hàm trả về nhiều giá trị (dùng tuple)
def min_max(numbers):
    return min(numbers), max(numbers)  # Trả về tuple

result = min_max([3, 1, 4, 1, 5])
print(f"Kết quả: {result}, kiểu: {type(result)}")

# Unpacking
lo, hi = min_max([3, 1, 4, 1, 5])
print(f"Min={lo}, Max={hi}")
```

## 3. Dictionary (Từ điển)

- **Key-Value pairs** (cặp khóa-giá trị)
- **Key** phải duy nhất và immutable (str, int, tuple)
- **Mutable** (thay đổi được)
- Kí hiệu: `{ }`

```python
# Tạo dictionary
student = {
    "name": "Hải",
    "age": 20,
    "gpa": 3.5,
    "courses": ["PFP", "PRO", "MAD"]
}

print(f"student: {student}")
print(f"Kiểu: {type(student)}")
```

```python
# Truy cập giá trị
print(f"Tên: {student['name']}")
print(f"GPA: {student['gpa']}")

# ❌ LỖI: Key không tồn tại
try:
    print(student['email'])  # KeyError!
except KeyError as e:
    print(f"LỖI: KeyError - {e}")

# ✅ Dùng get() - an toàn, không lỗi
print(f"Email: {student.get('email', 'Chưa có')}")
```

```python
# Thêm / Sửa / Xóa
d = {"a": 1, "b": 2}
print(f"Ban đầu: {d}")

d["c"] = 3            # Thêm key mới
print(f"Thêm c: {d}")

d["a"] = 99           # Sửa giá trị
print(f"Sửa a: {d}")

del d["b"]            # Xóa key
print(f"Xóa b: {d}")

val = d.pop("c")      # Xóa và trả về giá trị
print(f"pop('c'): {val}, dict: {d}")
```

```python
# Duyệt dictionary
scores = {"An": 8.5, "Bình": 7.0, "Chi": 9.5}

# Duyệt keys
print("Keys:", list(scores.keys()))

# Duyệt values
print("Values:", list(scores.values()))

# Duyệt cả key-value (KHUYÊN DÙNG)
print("\nBảng điểm:")
for name, score in scores.items():
    status = "Đạt" if score >= 5 else "Trượt"
    print(f"  {name}: {score} ({status})")
```

```python
# Kiểm tra key tồn tại
d = {"name": "Hải", "age": 20}

print(f"'name' in d: {'name' in d}")    # True
print(f"'email' in d: {'email' in d}")  # False
```

```python
# Ứng dụng: Đếm tần suất ký tự
s = "programming"
freq = {}
for char in s:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(f"Chuỗi: '{s}'")
print(f"Tần suất: {freq}")

# Tìm ký tự xuất hiện nhiều nhất
max_char = max(freq, key=freq.get)
print(f"Nhiều nhất: '{max_char}' ({freq[max_char]} lần)")
```

```python
# ❌ Key phải là immutable (không dùng list làm key)
try:
    d = {[1, 2]: "value"}  # TypeError!
except TypeError as e:
    print(f"LỖI: {e}")

# ✅ Dùng tuple làm key được
d = {(1, 2): "tọa độ"}
print(f"d[(1,2)] = {d[(1, 2)]}")
```

## 4. Set (Tập hợp)

- **Không trùng lặp**
- **Không có thứ tự** (unordered)
- **Mutable**
- Kí hiệu: `{ }` (nhưng `{}` rỗng là dict!)

```python
# Tạo set
s1 = {1, 2, 3, 3, 2, 1}  # Tự loại bỏ trùng
print(f"Set: {s1}")       # {1, 2, 3}

# ⚠️ Set rỗng phải dùng set(), KHÔNG phải {}
empty_set = set()
empty_dict = {}
print(f"set():  {type(empty_set)}")
print(f"{{}}:    {type(empty_dict)}")  # dict!
```

```python
# Loại bỏ trùng từ list
nums = [1, 2, 3, 2, 1, 4, 3, 5]
unique = list(set(nums))
print(f"Gốc: {nums}")
print(f"Không trùng: {unique}")
```

```python
# Phép toán tập hợp
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(f"A: {a}")
print(f"B: {b}")
print(f"Hợp (A | B):  {a | b}")     # {1,2,3,4,5,6}
print(f"Giao (A & B): {a & b}")      # {3, 4}
print(f"Hiệu (A - B): {a - b}")     # {1, 2}
print(f"Đối xứng (A ^ B): {a ^ b}") # {1, 2, 5, 6}
```

```python
# ❌ Set KHÔNG có index
s = {1, 2, 3}
try:
    print(s[0])  # TypeError!
except TypeError as e:
    print(f"LỖI: {e}")
    print("→ Set không có thứ tự, không truy cập bằng index")
```

## 5. So Sánh Các Cấu Trúc Dữ Liệu

| | List | Tuple | Dict | Set |
|---|---|---|---|---|
| Kí hiệu | `[]` | `()` | `{k:v}` | `{}` |
| Mutable | ✅ | ❌ | ✅ | ✅ |
| Thứ tự | ✅ | ✅ | ✅ (3.7+) | ❌ |
| Trùng lặp | ✅ | ✅ | Key: ❌ | ❌ |
| Index | ✅ | ✅ | Key | ❌ |
| Dùng khi | Dữ liệu thay đổi | Dữ liệu cố định | Ánh xạ key→value | Loại trùng |

```python
# Chuyển đổi giữa các kiểu
nums = [1, 2, 3, 2, 1]

t = tuple(nums)       # List → Tuple
s = set(nums)         # List → Set (loại trùng)
back = list(s)        # Set → List

print(f"List:  {nums}")
print(f"Tuple: {t}")
print(f"Set:   {s}")
print(f"Back:  {back}")
```

## 6. Nhập Dữ Liệu Từ Bàn Phím

```python
# Pattern đề thi: nhập list từ 1 dòng
# Input: "3 2 5 7 6"
line = "3 2 5 7 6"  # Giả lập input()
line = input("List")

numbers = list(map(int, line.split()))
print(f"List: {numbers}")

# Hoặc dùng comprehension
numbers = [int(x) for x in line.split()]
print(f"Comprehension: {numbers}")
```
