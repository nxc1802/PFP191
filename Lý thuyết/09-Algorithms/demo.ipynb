## 1. Kiểm Tra Số Nguyên Tố

```python
# Số nguyên tố: > 1 và chỉ chia hết cho 1 và chính nó
# Tối ưu: chỉ kiểm tra đến √n

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Test
print("Các số nguyên tố từ 1-30:")
for i in range(1, 31):
    if is_prime(i):
        print(i, end=" ")
```

```python
# ⚠️ Các edge case cần nhớ
print(f"is_prime(0): {is_prime(0)}")   # False
print(f"is_prime(1): {is_prime(1)}")   # False (1 KHÔNG phải nguyên tố)
print(f"is_prime(2): {is_prime(2)}")   # True (số nguyên tố nhỏ nhất)
print(f"is_prime(-5): {is_prime(-5)}") # False (số âm)
```

```python
# ❌ SAI: Kiểm tra không tối ưu
def is_prime_slow(n):
    if n < 2:
        return False
    for i in range(2, n):  # Kiểm tra đến n-1 → chậm!
        if n % i == 0:
            return False
    return True

# ✅ Tối ưu: chỉ cần kiểm tra đến √n
# Vì nếu n = a × b, thì a hoặc b phải ≤ √n
import time
n = 999983  # Số nguyên tố lớn

start = time.time()
is_prime_slow(n)
slow = time.time() - start

start = time.time()
is_prime(n)
fast = time.time() - start

print(f"Chậm (đến n): {slow:.6f}s")
print(f"Nhanh (đến √n): {fast:.6f}s")
print(f"Nhanh hơn: {slow/fast:.0f}x")
```

```python
# Tổng số nguyên tố < n (bài Q2 đề thi)
def sum_primes(n):
    total = 0
    for i in range(2, n):
        if is_prime(i):
            total += i
    return total

n = 10
primes = [i for i in range(2, n) if is_prime(i)]
print(f"Primes < {n}: {primes}")
print(f"Tổng: {' + '.join(map(str, primes))} = {sum_primes(n)}")
```

## 2. Tìm Min/Max

```python
# Min của 3 số (bài Q1 đề thi)
def min_of_three(a, b, c):
    if a <= b and a <= c:
        return a
    elif b <= a and b <= c:
        return b
    else:
        return c

print(f"min(3, 7, 6) = {min_of_three(3, 7, 6)}")
print(f"min(5, 5, 5) = {min_of_three(5, 5, 5)}")
print(f"min(-1, 0, 1) = {min_of_three(-1, 0, 1)}")
```

```python
# Min/Max trong list (không dùng hàm có sẵn)
def find_min(numbers):
    min_val = numbers[0]
    for n in numbers:
        if n < min_val:
            min_val = n
    return min_val

nums = [5, 2, 8, 1, 9, 3]
print(f"List: {nums}")
print(f"Min: {find_min(nums)}")

# ❌ LỖI nếu list rỗng
try:
    find_min([])
except IndexError as e:
    print(f"LỖI list rỗng: {e}")
```

## 3. Trung Bình Có Điều Kiện

```python
# TB các số chia hết cho x (bài Q3 đề thi)
def avg_divisible(numbers, x):
    divisible = [n for n in numbers if n % x == 0]
    if len(divisible) == 0:
        return 0
    return sum(divisible) / len(divisible)

nums = [3, 2, 5, 7, 6]
x = 3
div = [n for n in nums if n % x == 0]
print(f"List: {nums}, x = {x}")
print(f"Chia hết cho {x}: {div}")
print(f"TB: ({' + '.join(map(str, div))}) / {len(div)} = {avg_divisible(nums, x):.2f}")
```

```python
# ⚠️ Edge case: Không có số nào chia hết
nums = [1, 3, 5, 7]
x = 2
print(f"Không có số chẵn: {avg_divisible(nums, x):.2f}")

# ❌ Nếu quên check len == 0 → ZeroDivisionError!
```

## 4. Tổng Chữ Số Trong Chuỗi

```python
# Bài Q4 đề thi
def sum_digits(s):
    total = 0
    for char in s:
        if char.isdigit():
            total += int(char)
    return total

s = "abc123def45"
digits = [c for c in s if c.isdigit()]
print(f"Chuỗi: '{s}'")
print(f"Chữ số: {digits}")
print(f"Tổng: {' + '.join(digits)} = {sum_digits(s)}")
```

```python
# ⚠️ Lưu ý: int('2') = 2, KHÔNG PHẢI ord('2') = 50
char = '7'
print(f"int('{char}'): {int(char)}")     # 7 (giá trị số)
print(f"ord('{char}'): {ord(char)}")     # 55 (mã ASCII)
```

## 5. Giai Thừa & Fibonacci

```python
# Giai thừa: n! = n × (n-1) × ... × 1
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Đệ quy
def factorial_rec(n):
    if n <= 1: return 1
    return n * factorial_rec(n - 1)

for i in range(1, 8):
    print(f"{i}! = {factorial(i)}")
```

```python
# Fibonacci: 0, 1, 1, 2, 3, 5, 8, 13...
def fibonacci(n):
    if n <= 0: return []
    if n == 1: return [0]
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib

print(f"Fibonacci(10): {fibonacci(100)}")
```

## 6. Sắp Xếp

```python
# Bubble Sort (sắp xếp nổi bọt)
def bubble_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

nums = [64, 34, 25, 12, 22, 11, 90]
print(f"Trước: {nums}")
print(f"Sau:   {bubble_sort(nums)}")
```

```python
# Selection Sort (sắp xếp chọn)
def selection_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

nums = [64, 34, 25, 12, 22, 11, 90]
print(f"Selection sort: {selection_sort(nums)}")
```

```python
# ⚠️ sorted() vs .sort()
nums = [3, 1, 4, 1, 5]

# sorted() → list MỚI, gốc KHÔNG đổi
new = sorted(nums)
print(f"sorted: new={new}, gốc={nums}")

# .sort() → sửa TẠI CHỖ, trả về None
result = nums.sort()
print(f".sort(): result={result}, nums={nums}")
# result = None!
```
