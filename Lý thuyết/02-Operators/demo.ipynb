## 1. Toán Tử Số Học

| Toán tử | Mô tả | Ví dụ |
|---------|-------|-------|
| `+` | Cộng | `5 + 3 = 8` |
| `-` | Trừ | `5 - 3 = 2` |
| `*` | Nhân | `5 * 3 = 15` |
| `/` | Chia (float) | `5 / 3 = 1.667` |
| `//` | Chia lấy phần nguyên | `5 // 3 = 1` |
| `%` | Chia lấy dư (modulo) | `5 % 3 = 2` |
| `**` | Lũy thừa | `5 ** 3 = 125` |

```python
a = 10
b = 3.0

print(f"{a} + {b} = {a + b}")      # Cộng
print(f"{a} - {b} = {a - b}")      # Trừ
print(f"{a} * {b} = {a * b}")      # Nhân
print(f"{a} / {b} = {a / b:.4f}")  # Chia (luôn trả float)
print(f"{a} // {b} = {a // b}")    # Chia lấy phần nguyên
print(f"{a} % {b} = {a % b}")      # Chia lấy dư
print(f"{a} ** {b} = {a ** b}")    # Lũy thừa
```

```python
# ⚠️ Sự khác nhau giữa / và //
print("--- Phép chia / (luôn float) ---")
print(f"10 / 3 = {10 / 3}")    # 3.3333...
print(f"10 / 2 = {10 / 2}")    # 5.0 (vẫn là float!)
print(f"Kiểu: {type(10 / 2)}") # float

print("\n--- Phép chia // (lấy phần nguyên) ---")
print(f"10 // 3 = {10 // 3}")  # 3
print(f"10 // 2 = {10 // 2}")  # 5 (int)
print(f"Kiểu: {type(10 // 2)}") # int
```

```python
# ⚠️ Phép chia // với số ÂM - kết quả KHÔNG giống C/Java!
print(f"7 // 2 = {7 // 2}")      # 3 (bình thường)
print(f"-7 // 2 = {-7 // 2}")    # -4 (Python làm tròn XUỐNG)
print(f"7 // -2 = {7 // -2}")    # -4

# Giải thích: Python luôn làm tròn về phía -∞
# -7 / 2 = -3.5 → làm tròn xuống = -4
```

```python
# ⚠️ Modulo % - Rất hữu ích để kiểm tra chia hết
print("--- Ứng dụng modulo ---")
print(f"10 % 2 = {10 % 2}")  # 0 → 10 chia hết cho 2
print(f"10 % 3 = {10 % 3}")  # 1 → 10 KHÔNG chia hết cho 3
print(f"15 % 5 = {15 % 5}")  # 0 → 15 chia hết cho 5

# Kiểm tra số chẵn/lẻ
n = 7
if n % 2 == 0:
    print(f"{n} là số chẵn")
else:
    print(f"{n} là số lẻ")
```

```python
# ❌ LỖI: Chia cho 0
print(10 / 0)  # ZeroDivisionError: division by zero
```

```python
# ❌ LỖI: Chia lấy dư cho 0
print(10 % 0)  # ZeroDivisionError
```

```python
# ⚠️ Toán tử + và * với string
print("Hello" + " " + "World")  # Nối chuỗi: "Hello World"
print("Ha" * 3)                  # Lặp chuỗi: "HaHaHa"
print("-" * 30)                  # Dùng để tạo đường kẻ
```

```python
# ❌ LỖI: Không thể cộng string với int
print("Age: " + 20)  # TypeError: can only concatenate str to str
```

```python
# ✅ Cách sửa: Chuyển int sang str
print("Age: " + str(20))  # "Age: 20"
# Hoặc dùng f-string (tiện hơn)
print(f"Age: {20}")
a = 30
print(f"Age: {a}")
```

## 2. Toán Tử So Sánh

Trả về `True` hoặc `False`

| Toán tử | Mô tả |
|---------|-------|
| `==` | Bằng |
| `!=` | Khác |
| `>` | Lớn hơn |
| `<` | Nhỏ hơn |
| `>=` | Lớn hơn hoặc bằng |
| `<=` | Nhỏ hơn hoặc bằng |

```python
x = 10
y = 5

print(f"{x} == {y}: {x == y}")  # False
print(f"{x} != {y}: {x != y}")  # True
print(f"{x} > {y}: {x > y}")    # True
print(f"{x} < {y}: {x < y}")    # False
print(f"{x} >= {y}: {x >= y}")  # True
print(f"{x} <= {y}: {x <= y}")  # False
```

```python
# ⚠️ LỖI RẤT THƯỜNG GẶP: == vs =
# = là GÁN giá trị
# == là SO SÁNH

x = 10     # Gán: x nhận giá trị 5
x == 5    # So sánh: x có bằng 5 không? → True
print(f"x = 5 (gán) → x = {x}")
print(f"x == 5 (so sánh) → {x == 5}")
```

```python
# ⚠️ So sánh khác kiểu dữ liệu
print(f"5 == 5.0: {5 == 5.0}")     # True (giá trị bằng nhau)
print(f"5 == '5': {5 == '5'}")     # False (int vs str)
print(f"True == 1: {True == 1}")   # True (bool là con của int)
print(f"False == 0: {False == 0}") # True
```

```python
# ⚠️ So sánh string - theo thứ tự từ điển (ASCII)
print(f"'apple' < 'banana': {'apple' < 'banana'}")  # True (a < b)
print(f"'abc' < 'abd': {'abc' < 'abd'}")            # True (c < d)
print(f"'A' < 'a': {'A' < 'a'}")                    # True (A=65, a=97)
```

## 3. Toán Tử Logic

| Toán tử | Mô tả |
|---------|-------|
| `and` | VÀ - Cả hai True mới True |
| `or` | HOẶC - Ít nhất một True thì True |
| `not` | PHỦ ĐỊNH - Đảo giá trị |

```python
# Bảng chân trị (Truth Table)
print("--- AND ---")
print(f"True and True = {True and True}")    # True
print(f"True and False = {True and False}")  # False
print(f"False and True = {False and True}")  # False
print(f"False and False = {False and False}") # False

print("\n--- OR ---")
print(f"True or True = {True or True}")      # True
print(f"True or False = {True or False}")    # True
print(f"False or True = {False or True}")    # True
print(f"False or False = {False or False}")  # False

print("\n--- NOT ---")
print(f"not True = {not True}")   # False
print(f"not False = {not False}") # True
```

```python
# Ví dụ thực tế: Kiểm tra đăng nhập
username = "admin"
password = "12345"

is_valid = username == "admin" and password == "12345"
print(f"Đăng nhập hợp lệ: {is_valid}")

# Ví dụ: Kiểm tra tuổi vào quán bar
age = 17
has_id = True
can_enter = age >= 18 and has_id
print(f"Tuổi: {age}, Có CMND: {has_id} → Được vào: {can_enter}")
```

```python
# ⚠️ Short-circuit evaluation (đánh giá tắt)
# and: Nếu vế trái False → KHÔNG tính vế phải
# or: Nếu vế trái True → KHÔNG tính vế phải

# Ví dụ: Tránh chia cho 0
x = 0
# Nếu x == 0, Python sẽ KHÔNG tính 10/x (tránh lỗi)
result = x != 0 and 10 / x > 2
print(f"x = {x}, result = {result}")  # False (không lỗi!)
```

## 4. Toán Tử Gán

| Toán tử | Tương đương |
|---------|-------------|
| `x += n` | `x = x + n` |
| `x -= n` | `x = x - n` |
| `x *= n` | `x = x * n` |
| `x /= n` | `x = x / n` |
| `x //= n` | `x = x // n` |
| `x %= n` | `x = x % n` |
| `x **= n` | `x = x ** n` |

```python
x = 10
print(f"x ban đầu = {x}")

x += 5    # x = x + 5 = 15
print(f"x += 5 → {x}")

x -= 3    # x = 15 - 3 = 12
print(f"x -= 3 → {x}")

x *= 2    # x = 12 * 2 = 24
print(f"x *= 2 → {x}")

x //= 5   # x = 24 // 5 = 4
print(f"x //= 5 → {x}")

x **= 3   # x = 4 ** 3 = 64
print(f"x **= 3 → {x}")
```

## 5. Độ Ưu Tiên Toán Tử

Từ **cao** đến **thấp**:
1. `**` (lũy thừa)
2. `*`, `/`, `//`, `%` (nhân, chia)
3. `+`, `-` (cộng, trừ)
4. `==`, `!=`, `<`, `>`, `<=`, `>=` (so sánh)
5. `not` (phủ định)
6. `and` (và)
7. `or` (hoặc)

💡 **Mẹo**: Khi không chắc chắn → **DÙNG NGOẶC `()`**

```python
# Ví dụ về độ ưu tiên
print(f"2 + 3 * 4 = {2 + 3 * 4}")        # 14 (nhân trước)
print(f"(2 + 3) * 4 = {(2 + 3) * 4}")    # 20 (ngoặc trước)
print(f"2 ** 3 ** 2 = {2 ** 3 ** 2}")     # 512 (** tính từ PHẢI qua TRÁI)
# 2 ** 3 ** 2 = 2 ** 9 = 512 (KHÔNG phải 8 ** 2 = 64)
```

```python
# ⚠️ Bẫy thường gặp: ** kết hợp phải
print(f"2 ** 3 ** 2 = {2 ** 3 ** 2}")     # = 2 ** 9 = 512
print(f"(2 ** 3) ** 2 = {(2 ** 3) ** 2}") # = 8 ** 2 = 64
# Nếu muốn tính (2^3)^2 → phải dùng ngoặc!
```

```python
# Ví dụ tổng hợp phức tạp
result = 2 ** 3 + 4 * 5 - 10 // 3
print(f"2 ** 3 + 4 * 5 - 10 // 3")
print(f"= 8 + 20 - 3")
print(f"= {result}")

# Logic kết hợp
x = 5
result = x > 3 and x < 10 or x == 0
print(f"\nx = {x}")
print(f"x > 3 and x < 10 or x == 0")
print(f"= True and True or False")
print(f"= True or False")
print(f"= {result}")
```

```python
# Bài 1: Tách các chữ số của một số 3 chữ số
num = 365

hundreds = num // 100        # 3
tens = (num // 10) % 10      # 6
ones = num % 10              # 5

print(f"Số: {num}")
print(f"Hàng trăm: {hundreds}")
print(f"Hàng chục: {tens}")
print(f"Hàng đơn vị: {ones}")
```
