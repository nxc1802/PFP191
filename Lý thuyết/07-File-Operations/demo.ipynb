## 1. Mở File - `open()`

```python
f = open("tên_file", "mode")
```

| Mode | Mô tả |
|------|--------|
| `'r'` | **Read** - Đọc (mặc định). Lỗi nếu file không tồn tại |
| `'w'` | **Write** - Ghi. Tạo mới hoặc **XÓA SẠCH** nội dung cũ |
| `'a'` | **Append** - Nối thêm vào cuối file |
| `'r+'` | **Read+Write** - Đọc và ghi |
| `'x'` | **Create** - Tạo mới. Lỗi nếu file đã tồn tại |

````python
# Tạo file mẫu để demo
with open("sample.txt", "w") as f:
    f.write("Dòng 1: Hello Python\n")
    f.write("Dòng 2: File Operations\n")
    f.write("Dòng 3: PFP Course\n")
    f.write("Dòng 4: 2024\n")
    f.write("Dòng 5: The End")

print("✅ Đã tạo file sample.txt")
````

---
## 2. Đọc File

````python
# Cách 1: read() - đọc TOÀN BỘ file thành 1 string
with open("sample.txt", "r") as f:
    content = f.read()

print("=== read() ===")
print(content)
print(f"Kiểu: {type(content)}")
````

````python
# Cách 2: readline() - đọc TỪNG DÒNG MỘT
print("=== readline() ===")
with open("sample.txt", "r") as f:
    line1 = f.readline()  # Đọc dòng 1
    line2 = f.readline()  # Đọc dòng 2
    print(f"Dòng 1: {line1}", end="")  # Đã có \n
    print(f"Dòng 2: {line2}", end="")
````

````python
# Cách 3: readlines() - đọc tất cả thành LIST
print("=== readlines() ===")
with open("sample.txt", "r") as f:
    lines = f.readlines()

print(f"Kiểu: {type(lines)}")
print(f"Số dòng: {len(lines)}")
print(f"Nội dung: {lines}")
# ⚠️ Mỗi dòng kết thúc bằng \n
````

````python
# ✅ Cách tốt nhất: Lặp trực tiếp (tiết kiệm bộ nhớ)
print("=== for line in f ===")
with open("sample.txt", "r") as f:
    for i, line in enumerate(f, 1):
        print(f"  [{i}] {line.strip()}")  # strip() bỏ \n
````

````python
# ⚠️ strip() vs không strip()
with open("sample.txt", "r") as f:
    for line in f:
        print(f"  Không strip: '{line}'")  # Có \n ở cuối
        break

with open("sample.txt", "r") as f:
    for line in f:
        print(f"  Có strip:    '{line.strip()}'")  # Sạch sẽ ✓
        break
````

## 3. Ghi File

````python
# Mode 'w' - GHI ĐÈ (⚠️ xóa sạch nội dung cũ!)
with open("output.txt", "w") as f:
    f.write("Dòng 1\n")
    f.write("Dòng 2\n")

# Kiểm tra
with open("output.txt", "r") as f:
    print("Sau 'w':")
    print(f.read())
````

````python
# Mode 'a' - NỐI THÊM (giữ nguyên nội dung cũ)
with open("output.txt", "a") as f:
    f.write("Dòng 3 (nối thêm)\n")

with open("output.txt", "r") as f:
    print("Sau 'a':")
    print(f.read())
````

````python
# ⚠️ CẢNH BÁO: Mode 'w' XÓA SẠCH file cũ!
with open("output.txt", "w") as f:
    f.write("Chỉ còn dòng này\n")

with open("output.txt", "r") as f:
    print("Sau 'w' lần 2:")
    print(f.read())
    print("→ Dòng 1, 2, 3 đã bị XÓA HẾT!")
````

````python
# writelines() - ghi list các dòng
lines = ["Apple\n", "Banana\n", "Cherry\n"]

with open("fruits.txt", "w") as f:
    f.writelines(lines)

with open("fruits.txt", "r") as f:
    print(f.read())

# ⚠️ writelines() KHÔNG tự thêm \n!
# Phải tự thêm \n vào mỗi phần tử
````

## 4. `with` vs Không `with`

````python
# ❌ Không dùng with - phải tự đóng file
f = open("sample.txt", "r")
content = f.read()
f.close()  # PHẢI đóng! Quên → rò rỉ tài nguyên

# ✅ Dùng with - TỰ ĐỘNG đóng file
with open("sample.txt", "r") as f:
    content = f.read()
# File tự đóng khi ra khỏi with, KỂ CẢ KHI CÓ LỖI

print("→ Luôn dùng 'with' cho an toàn!")
````

````python
# ❌ LỖI: Đọc file sau khi đóng
with open("sample.txt", "r") as f:
    pass  # File đóng sau dòng này

try:
    content = f.read()  # ValueError: file đã đóng!
except ValueError as e:
    print(f"LỖI: {e}")
````

## 5. Xử Lý Lỗi File

````python
# ❌ LỖI: File không tồn tại
try:
    with open("khong_ton_tai.txt", "r") as f:
        content = f.read()
except FileNotFoundError:
    print("❌ FileNotFoundError: File không tồn tại!")
````

````python
# ✅ Đọc file an toàn
import os

filename = "sample.txt"

# Cách 1: try-except
try:
    with open(filename, "r") as f:
        print(f.read())
except FileNotFoundError:
    print(f"File '{filename}' không tồn tại")

# Cách 2: Kiểm tra trước
if os.path.exists(filename):
    print(f"File '{filename}' tồn tại ✓")
else:
    print(f"File '{filename}' không tồn tại")
````

## 6. Ứng Dụng Thực Tế

````python
# Bài 1: Đọc file và tính điểm trung bình
# Tạo file điểm
with open("scores.txt", "w") as f:
    f.write("An 8.5\n")
    f.write("Bình 7.0\n")
    f.write("Chi 9.5\n")
    f.write("Dung 6.0\n")

# Đọc và xử lý
total = 0
count = 0
with open("scores.txt", "r") as f:
    for line in f:
        parts = line.strip().split()
        name = parts[0]
        score = float(parts[1])
        print(f"  {name}: {score}")
        total += score
        count += 1

print(f"\nĐiểm trung bình: {total/count:.2f}")
````

````python
# Bài 2: Đếm từ trong file
with open("sample.txt", "r") as f:
    content = f.read()

words = content.split()
lines = content.strip().split("\n")
chars = len(content)

print(f"Số dòng: {len(lines)}")
print(f"Số từ:   {len(words)}")
print(f"Số ký tự: {chars}")
````

````python
# Bài 3: Copy file (lọc dòng trống)
# Tạo file có dòng trống
with open("input.txt", "w") as f:
    f.write("Hello\n\nWorld\n\n\nPython\n")

# Copy bỏ dòng trống
with open("input.txt", "r") as fin:
    with open("output_clean.txt", "w") as fout:
        for line in fin:
            if line.strip():  # Bỏ qua dòng trống
                fout.write(line)

# Kiểm tra
with open("output_clean.txt", "r") as f:
    print("File đã lọc:")
    print(f.read())
````

````python
# Dọn dẹp file demo
import os
for f in ["sample.txt", "output.txt", "fruits.txt", "scores.txt", "input.txt", "output_clean.txt"]:
    if os.path.exists(f):
        os.remove(f)
print("🧹 Đã dọn dẹp file demo")
````
