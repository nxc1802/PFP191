## 1. Tạo Chuỗi

```python
s1 = 'Nháy đơn'
s2 = "Nháy kép"
s3 = '''Chuỗi
            nhiều dòng'''

print(s1)
print(s2)
print(s3)
```

```python
# Escape characters
print("Dòng 1\nDòng 2")      # \n = xuống dòng
print("Col1\tCol2")           # \t = tab
print("Dấu nháy: \"Hi\"")   # \" = dấu nháy kép
```

---
## 2. Truy Cập & Immutability

```python
s = "Python"
print(f"s[0]: '{s[0]}'")
print(f"s[-1]: '{s[-2]}'")
print(f"s[0:3]: '{s[0:3]}'")
print(f"s[::-1]: '{s[::-1]}'")
```

```python
# ❌ LỖI: String KHÔNG thể thay đổi
s = "Hello"
try:
    s[0] = "h"  # TypeError!
except TypeError as e:
    print(f"LỖI: {e}")

# ✅ Tạo string mới
s = "h" + s[1:]
print(f"Cách đúng: {s}")
```

---
## 3. Phương Thức Quan Trọng

```python
s = "Hello World"
print(f"upper():   '{s.upper()}'")
print(f"lower():   '{s.lower()}'")
print(f"title():   '{s.title()}'")

# ⚠️ Trả về string MỚI, KHÔNG thay đổi gốc
s.upper()
print(f"Sau s.upper(): '{s}'")  # Vẫn 'Hello World'!
```

```python
# Tìm kiếm
s = "Hello World"
print(f"find('o'):    {s.find('o')}")       # 4
print(f"find('xyz'):  {s.find('xyz')}")     # -1
print(f"count('o'):   {s.count('o')}")      # 2
print(f"'World' in s: {'World' in s}")      # True
```

```python
# ⚠️ find() vs index()
s = "Hello"
print(f"find('xyz'): {s.find('xyz')}")  # -1 (an toàn)

try:
    s.index('xyz')  # ValueError!
except ValueError as e:
    print(f"index('xyz'): LỖI - {e}")
```

```python
# strip, replace, split, join
print(f"strip: '{'                      Hi  '.strip()}'")
print(f"replace: {'Hello'.replace('l', 'L')}")

parts = "a,b,c".split(",")
print(f"split: {parts}")
print(f"join: {' - '.join(parts)}")
```

## 4. Kiểm Tra Ký Tự

```python
print(f"'123'.isdigit():  {'123'.isdigit()}")    # True
print(f"'12.3'.isdigit(): {'12.3'.isdigit()}")   # False (dấu .)
print(f"'abc'.isalpha():  {'abc'.isalpha()}")    # True
print(f"''.isdigit():     {''.isalpha()}")       # False (rỗng)
```

```python
# Phân tích từng ký tự
s = "PFP PE SU23"
print(f"Chuỗi: '{s}'\n")
for char in s:
    if char.isdigit():
        print(f"  '{char}' → SỐ (giá trị: {int(char)})")
    elif char.isalpha():
        print(f"  '{char}' → CHỮ CÁI")
    else:
        print(f"  '{char}' → KHOẢNG TRẮNG")
```

## 5. f-string Formatting

```python
gpa = 3.567
print(f"2 chữ số: {gpa:.2f}")
print(f"Số lớn: {1000000:,}")
print(f"Phần trăm: {0.856:.1%}")
print(f"{'Trái':<15}|")
print(f"{'Phải':>15}|")
```

## 6. Pattern Đề Thi

```python
# Tổng chữ số trong chuỗi
def sum_digits(s):
    total = 0
    for char in s:
        if char.isdigit():
            total += int(char)
    return total

print(f"sum_digits('abc123'): {sum_digits('abc123')}")  # 6
```

```python
# Đếm từ (split tự bỏ space thừa)
s = "  Python   is   great  "
print(s.split())
print(f"Số từ: {len(s.split())}")
```

```python
# Đảo chuỗi / Palindrome
print(f"'Python'[::-1] = {'Python'[::-1]}")

def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

for w in ["radar", "hello", "level"]:
    print(f"  '{w}' → {'✓' if is_palindrome(w) else '✗'}")
```

```python
# ⚠️ Cộng string vs Cộng số
print("'1' + '2' =", '1' + '2')  # '12' (nối chuỗi)
print("1 + 2 =", int('1') + int('2'))            # 3

# ❌ TypeError khi cộng khác kiểu
try:
    print("5" + 3)
except TypeError as e:
    print(f"LỖI: {e}")
    print("✅ Cách sửa: int('5') + 3 =", int('5') + 3)
```
