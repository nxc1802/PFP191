## 1. Câu Lệnh `if` Cơ Bản

````python
# if đơn giản
age = 20

if age >= 18:
    print("Bạn đủ tuổi trưởng thành")
````

````python
# if-else
n = 7

if n % 2 == 0:
    print(f"{n} là số chẵn")
else:
    print(f"{n} là số lẻ")
````

````python
# if-elif-else: Xếp loại điểm
score = 75

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Điểm: {score} → Xếp loại: {grade}")
````

## 2. Các Lỗi Thường Gặp Với `if`

````python
# ❌ LỖI: Quên dấu hai chấm :
if True
    print("Hello")  # SyntaxError: expected ':'
````

````python
# ❌ LỖI: Thụt lề (indentation) sai
if True:
print("Hello")  # IndentationError: expected an indented block
````

````python
# ❌ LỖI LOGIC: Dùng = thay vì ==
x = 5

# Sai: if x = 5 → SyntaxError (= là gán, KHÔNG phải so sánh)
# Đúng:
if x == 5:
    print("x bằng 5")
````

````python
# ⚠️ Thứ tự elif QUAN TRỌNG!
score = 95

# ❌ Sai: Điều kiện RỘNG đặt trước → luôn vào đó
if score >= 60:
    print(f"→ D (SAI! lẽ ra phải là A)")  # Luôn vào đây!
elif score >= 90:
    print(f"→ A (không bao giờ đến được)")
````

````python
# ✅ Đúng: Điều kiện HẸP (khó thỏa mãn hơn) đặt TRƯỚC
score = 95

if score >= 90:
    print("→ A ✓")  # Vào đúng!
elif score >= 80:
    print("→ B")
elif score >= 60:
    print("→ D")
````

````python
# if lồng nhau (nested if)
age = 20
has_ticket = True

if age >= 18:
    if has_ticket:
        print("Được vào rạp")
    else:
        print("Cần mua vé")
else:
    print("Chưa đủ tuổi")

# Viết gọn hơn bằng and:
if age >= 18 and has_ticket:
    print("→ Viết gọn: Được vào rạp")
````

## 3. Toán Tử 3 Ngôi (Ternary Operator)

Viết `if-else` trên **MỘT DÒNG**

```python
kết_quả = giá_trị_true if điều_kiện else giá_trị_false
```

````python
# Cách thường (3 dòng)
age = 20
if age >= 18:
    status = "Trưởng thành"
else:
    status = "Chưa trưởng thành"
print(f"Cách thường: {status}")

# Ternary operator (1 dòng) ✅
status = "Trưởng thành" if age >= 18 else "Chưa trưởng thành"
print(f"Ternary: {status}")
````

````python
# Ví dụ: Kiểm tra chẵn/lẻ
n = 7
result = "chẵn" if n % 2 == 0 else "lẻ"
print(f"{n} là số {result}")
````

````python
# Dùng trong print
x = 10
print(f"{x} là số {'dương' if x > 0 else 'không dương'}")
````

````python
# Dùng trong list comprehension
numbers = [1, -2, 3, -4, 5]
labels = ["dương" if n > 0 else "âm" for n in numbers]
print(f"Số: {numbers}")
print(f"Nhãn: {labels}")
````

````python
# Ternary lồng nhau (nested) - ĐỌC KHÓ, nên hạn chế!
n = 0
result = "dương" if n > 0 else ("âm" if n < 0 else "không")
print(f"{n} là số {result}")

# ⚠️ Nếu phức tạp → dùng if-elif-else cho dễ đọc!
````

````python
# Ternary để tìm min/max nhanh
a, b = 10, 20
min_val = a if a < b else b
max_val = a if a > b else b
print(f"a={a}, b={b}")
print(f"Min: {min_val}")
print(f"Max: {max_val}")
````

````python
# ❌ LỖI: Ternary PHẢI CÓ else
# result = "chẵn" if n % 2 == 0  # SyntaxError: thiếu else!

# ✅ Luôn phải có tối thiểu: giá_trị if điều_kiện else giá_trị
result = "chẵn" if 4 % 2 == 0 else "lẻ"
print(result)
````

## 4. Xử Lý Lỗi: `try` / `except`

Khi code có thể gây lỗi → dùng `try-except` để **bắt lỗi** thay vì crash chương trình.

```python
try:
    # Code có thể gây lỗi
except LoạiLỗi:
    # Xử lý khi lỗi xảy ra
```

````python
result = 10 / 0
print(f"Kết quả: {result}")
print("Chương trình tiếp tục chạy")
````

````python
# Ví dụ cơ bản: Chia cho 0

# ❌ Không bắt lỗi → chương trình crash
# print(10 / 0)  # ZeroDivisionError!

# ✅ Bắt lỗi bằng try-except
try:
    result = 10 / 0
    print(f"Kết quả: {result}")
except ZeroDivisionError:
    print("❌ Lỗi: Không thể chia cho 0!")

print("Chương trình vẫn tiếp tục chạy ✓")
````

````python
# Bắt lỗi chuyển đổi kiểu
inputs = ["123", "abc", "45.6", ""]

for s in inputs:
    try:
        n = int(s)
        print(f"  int('{s}') = {n} ✓")
    except ValueError:
        print(f"  int('{s}') → ❌ ValueError: không phải số nguyên!")
````

````python
# Bắt lỗi truy cập list
nums = [10, 20, 30]

for i in [0, 1, 5, -1]:
    try:
        print(f"  nums[{i}] = {nums[i]}")
    except IndexError:
        print(f"  nums[{i}] → ❌ IndexError: index ngoài phạm vi!")
````

### 4.1 Bắt Nhiều Loại Lỗi

````python
# Bắt nhiều loại lỗi khác nhau
test_cases = [("10", "2"), ("10", "0"), ("abc", "2")]

for a_str, b_str in test_cases:
    try:
        a = int(a_str)
        b = int(b_str)
        result = a / b
        print(f"  {a_str}/{b_str} = {result} ✓")
    except ValueError:
        print(f"  {a_str}/{b_str} → ❌ ValueError: không phải số!")
    except ZeroDivisionError:
        print(f"  {a_str}/{b_str} → ❌ ZeroDivisionError: chia cho 0!")
````

````python
# Bắt lỗi chung (Exception) - dùng khi không biết lỗi gì
try:
    result = int("hello") / 0
except Exception as e:
    print(f"Loại lỗi: {type(e).__name__}")
    print(f"Nội dung: {e}")

# ⚠️ Nên bắt lỗi CỤ THỂ thay vì Exception chung!
````

### 4.2 `try-except-else-finally`

````python
# Cấu trúc đầy đủ
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("  ❌ except: Chia cho 0!")  # Chạy khi CÓ lỗi
    else:
        print(f"  ✅ else: Kết quả = {result}")  # Chạy khi KHÔNG có lỗi
    finally:
        print("  🔚 finally: Luôn chạy dù có lỗi hay không")  # LUÔN chạy

print("=== Chia thành công ===")
divide(10, 3)

print("\n=== Chia cho 0 ===")
divide(10, 0)
````

### 4.3 Các Loại Lỗi Thường Gặp Trong PFP

| Loại lỗi | Khi nào | Ví dụ |
|----------|---------|-------|
| `ValueError` | Giá trị không hợp lệ | `int("abc")` |

| `NameError` | Biến chưa khai báo | `print(xyz)` |

### 4.4 Ứng Dụng: Nhập Liệu An Toàn

````python
# Hàm nhập số an toàn (không crash khi nhập sai)
def input_int_safe(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("⚠️ Vui lòng nhập số nguyên!")

# Demo: Uncomment để chạy thử
# n = input_int_safe("Nhập một số: ")
# print(f"Bạn đã nhập: {n}")

# Giả lập kết quả:
print("Nhập 'abc' → ⚠️ Vui lòng nhập số nguyên!")
print("Nhập '3.5' → ⚠️ Vui lòng nhập số nguyên!")
print("Nhập '42'  → Bạn đã nhập: 42 ✓")
````

````python
# Ứng dụng: Phép chia an toàn
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None  # Hoặc return 0, tuỳ yêu cầu

print(f"10 / 3 = {safe_divide(10, 3):.2f}")
print(f"10 / 0 = {safe_divide(10, 0)}")
````

## 5. Bài Tập Tổng Hợp

````python
# Bài 1: Xếp loại BMI
weight = 70  # kg
height = 1.75  # m
bmi = weight / height ** 2

# Dùng if-elif-else
if bmi < 18.5:
    category = "Thiếu cân"
elif bmi < 25:
    category = "Bình thường"
elif bmi < 30:
    category = "Thừa cân"
else:
    category = "Béo phì"

print(f"BMI: {bmi:.1f} → {category}")
````
