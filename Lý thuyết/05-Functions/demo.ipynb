````python
# Hàm đơn giản nhất
def say_hello():
    print("Hello, World!")
    print("123")

# Gọi hàm
say_hello()
say_hello()  # Có thể gọi nhiều lần
````

````python
# Hàm có tham số
def greet(name):
    print(f"Xin chào, {name}!")

greet("Hải")
greet("Lan")
````

````python
# ❌ LỖI: Gọi hàm trước khi định nghĩa
try:
    say_goodbye()  # NameError: name 'say_goodbye' is not defined
except NameError as e:
    print(f"LỖI: {e}")

def say_goodbye():
    print("Goodbye!")

# ✅ Giờ mới gọi được
say_goodbye()
````

````python
# ❌ LỖI: Số đối số không khớp
def add(a, b):
    return a + b

try:
    add(1)       # TypeError: thiếu đối số
except TypeError as e:
    print(f"LỖI 1: {e}")

try:
    add(1, 2, 3)  # TypeError: thừa đối số
except TypeError as e:
    print(f"LỖI 2: {e}")

# ✅ Đúng:
print(f"add(1, 2) = {add(1, 2)}")
````

## 2. Tham Số Mặc Định & Keyword Arguments

````python
# Tham số mặc định
def greet(name, greeting="Xin chào"):
    print(f"{greeting}, {name}!")

greet("Hải")                  # Dùng giá trị mặc định
greet("Hải", "Hello")         # Ghi đè giá trị mặc định
````

````python
# ❌ LỖI: Tham số mặc định phải đặt SAU tham số bình thường
# def func(a="default", b):  # SyntaxError!
#     pass

# ✅ Đúng:
def func(b, a="default"):
    print(f"a={a}, b={b}")

func(10)
func(10, "custom")
````

````python
# Keyword arguments - đặt tên cho đối số
def info(name, age, city):
    print(f"{name}, {age} tuổi, ở {city}")

# Có thể đảo thứ tự khi dùng keyword
info(age=20, city="Hà Nội", name="Hải")
````

````python
# ⚠️ BẪY: Tham số mặc định mutable (list, dict)
def add_item(item, lst=[]):
    lst.append(item)
    return lst

# Lần 1
print(add_item("a"))  # ['a']
# Lần 2 - ⚠️ Bất ngờ!
print(add_item("b"))  # ['a', 'b'] ← vẫn giữ 'a' từ lần trước!
print(add_item("c"))  # ['a', 'b', 'c']

# → List mặc định được CHIA SẺ giữa các lần gọi!
````

````python
# ✅ Cách đúng: Dùng None làm giá trị mặc định
def add_item_fixed(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add_item_fixed("a"))  # ['a']
print(add_item_fixed("b"))  # ['b'] ← Giờ đúng rồi!
````

## 3. Return

````python
# Hàm trả về giá trị
def square(n):
    return n ** 2

result = square(5)
print(f"5^2 = {result}")
````

````python
# Trả về nhiều giá trị (tuple)
def min_max(numbers):
    return min(numbers), max(numbers)

minimum, maximum = min_max([3, 1, 4, 1, 5, 9])
print(f"Min: {minimum}, Max: {maximum}")
````

````python
# ⚠️ Hàm KHÔNG CÓ return → trả về None
def say_hello():
    print("Hello!")
    # Không có return

result = say_hello()
print(f"Giá trị trả về: {result}")   # None
print(f"Kiểu: {type(result)}")        # NoneType
````

````python
# ⚠️ return kết thúc hàm NGAY LẬP TỨC
def demo():
    print("Dòng 1")
    return "done"
    print("Dòng 2")  # ← KHÔNG BAO GIỜ CHẠY!

result = demo()
print(f"Kết quả: {result}")
````

````python
# ⚠️ Nhầm lẫn print() vs return
def add_print(a, b):
    print(a + b)    # Chỉ IN ra, không trả về giá trị

def add_return(a, b):
    return a + b    # TRẢ VỀ giá trị

# Thử dùng kết quả
r1 = add_print(3, 5)   # In ra 8, nhưng r1 = None
r2 = add_return(3, 5)  # Không in gì, r2 = 8

print(f"\nr1 (print): {r1}")    # None
print(f"r2 (return): {r2}")    # 8

# r1 + 10 → TypeError vì None + 10 không được!
# r2 + 10 → 18 ✓
````

## 4. Phạm Vi Biến (Variable Scope)

````python
# Biến CỤC BỘ (local) - chỉ tồn tại trong hàm
def my_func():
    x = 100  # Biến cục bộ
    print(f"Trong hàm: x = {x}")

my_func()

try:
    print(x)  # ❌ NameError: x không tồn tại bên ngoài
except NameError as e:
    print(f"Ngoài hàm: LỖI - {e}")
````

````python
# Biến TOÀN CỤC (global)
x = 10  # Biến toàn cục

def read_global():
    print(f"Đọc biến toàn cục: x = {x}")  # ✅ Có thể ĐỌC

def try_modify():
    # ⚠️ Tạo biến CỤC BỘ mới tên x, KHÔNG thay đổi x toàn cục
    x = 999
    print(f"Trong hàm: x = {x}")

read_global()
try_modify()
print(f"Sau khi gọi hàm: x = {x}")  # Vẫn = 10!
````

````python
# Dùng 'global' để SỬA biến toàn cục
counter = 0

def increment():
    global counter  # Khai báo dùng biến toàn cục
    counter += 1

print(f"Trước: counter = {counter}")
increment()
increment()
increment()
print(f"Sau 3 lần gọi: counter = {counter}")  # 3

# ⚠️ Tránh dùng global quá nhiều! Dùng return thay thế
````

## 5. Hàm Đệ Quy (Recursion)

````python
# Giai thừa: n! = n × (n-1) × ... × 1
def factorial(n):
    if n <= 1:       # Base case (điều kiện dừng)
        return 1
    return n * factorial(n - 1)  # Recursive case

print(f"{5}! = {factorial(5)}")
````

factorial(5) = 5 * factorial(4) = 5 * 4 * factorial(3) = 5 * 4 * 3 * factorial(2) =. 5 * 4 * 3 * 2 * factorial(1) = 5 * 4 * 3 * 2 * 1

````python
# ❌ LỖI: Đệ quy không có điều kiện dừng
def infinite_recursion(n):
    return infinite_recursion(n + 1)  # Không bao giờ dừng!

try:
    infinite_recursion(1)
except RecursionError as e:
    print(f"LỖI: {e}")
    print("→ Python giới hạn ~1000 lần đệ quy")
````

## 6. Lambda Functions

Hàm **ẩn danh**, viết trên **một dòng**.  
Cú pháp: `lambda tham_số: biểu_thức`

````python
# Lambda cơ bản
square = lambda x: x ** 2
add = lambda a, b: a + b

print(f"square(5) = {square(5)}")
print(f"add(3, 4) = {add(3, 4)}")
````

````python
# Lambda hữu ích nhất khi dùng với sorted, map, filter
students = [("An", 8.5), ("Bình", 9.0), ("Chi", 7.5)]

# Sắp xếp theo điểm (phần tử thứ 2)
sorted_by_score = sorted(students, key=lambda s: s[1])
print(f"Tăng dần: {sorted_by_score}")

sorted_desc = sorted(students, key=lambda s: -s[1])
print(f"Giảm dần: {sorted_desc}")
````

## 7. Các Hàm Hay Dùng Trong Đề Thi

````python
# Hàm kiểm tra số nguyên tố
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# Test
for i in [1, 2, 3, 4, 5, 7, 11, 15, 17]:
    print(f"{i:2d} → {'Nguyên tố ✓' if is_prime(i) else 'Không'}")
````

````python
# Hàm tìm min trong 3 số (không dùng min())
def find_min(a, b, c):
    if a <= b and a <= c:
        return a
    elif b <= a and b <= c:
        return b
    else:
        return c

print(f"Min(3, 7, 6) = {find_min(3, 7, 6)}")
print(f"Min(10, 2, 5) = {find_min(10, 2, 5)}")
````

````python
# Hàm tính trung bình các phần tử chia hết cho x
def avg_divisible(numbers, x):
    divisible = [n for n in numbers if n % x == 0]
    if len(divisible) == 0:
        return 0
    return sum(divisible) / len(divisible)

nums = [3, 2, 5, 7, 6]
result = avg_divisible(nums, 3)
print(f"TB các số chia hết cho 3: {result:.2f}")
````
