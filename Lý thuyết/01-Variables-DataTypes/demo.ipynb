## 1. Khai Báo Biến

Biến là nơi **lưu trữ dữ liệu** trong chương trình.  
Python là ngôn ngữ **dynamic typing** → không cần khai báo kiểu trước.

```python
# Khai báo biến đơn giản
name = "Nguyen Van A"
age = 20
height = 1.75
is_student = True

print(f"Tên: {name}")
print(f"Tuổi: {age}")
print(f"Chiều cao: {height}m")
print(f"Là sinh viên: {is_student}")
```

### Quy tắc đặt tên biến

| ✅ Hợp lệ | ❌ Không hợp lệ | Lý do |
|-----------|----------------|-------|
| `my_name` | `2name` | Không bắt đầu bằng số |
| `_count` | `my-name` | Không dùng dấu gạch ngang |
| `age2` | `my name` | Không dùng khoảng trắng |
| `Name` | `for` | Không dùng từ khoá Python |

```python
# ✅ Tên biến hợp lệ
my_name = "Hải"
_count = 10
age2 = 25

print(my_name, _count, age2)
```

```python
# ❌ LỖI: Tên biến bắt đầu bằng số
2name = "Hải"  # SyntaxError
```

```python
# ❌ LỖI: Dùng từ khoá Python làm tên biến
for = 10  # SyntaxError: 'for' là từ khoá
```

```python
# ⚠️ Python phân biệt chữ hoa/thường
Name = "Hải"
name = "Lan"
NAME = "Minh"

print(f"Name = {Name}")
print(f"name = {name}")
print(f"NAME = {NAME}")
# → Ba biến KHÁC NHAU!
```

## 2. Kiểu Dữ Liệu Cơ Bản

| Kiểu | Mô tả | Ví dụ |
|------|-------|-------|
| `int` | Số nguyên | `10`, `-5`, `0` |
| `float` | Số thực (có phần thập phân) | `3.14`, `-2.5` |
| `str` | Chuỗi ký tự | `"Hello"`, `'Python'` |
| `bool` | Giá trị logic | `True`, `False` |

```python
# Kiểm tra kiểu dữ liệu bằng type()
a = 10
b = 3.14
c = "Hello"
d = True

print(f"a = {a}, kiểu: {type(a)}")
print(f"b = {b}, kiểu: {type(b)}")
print(f"c = {c}, kiểu: {type(c)}")
print(f"d = {d}, kiểu: {type(d)}")
```

```python
# ⚠️ Chú ý: Phép chia luôn trả về float
x = 10
y = 2
result = x / y

print(f"{x} / {y} = {result}")
print(f"Kiểu: {type(result)}")  # float, không phải int!
```

```python
# ⚠️ bool là kiểu con của int
print(f"True + True = {True + True}")    # 2
print(f"True + False = {True + False}")  # 1
print(f"True * 5 = {True * 5}")          # 5
print(f"isinstance(True, int) = {isinstance(True, int)}")  # True!
```

## 3. Chuyển Đổi Kiểu Dữ Liệu (Type Casting)

Dùng các hàm: `int()`, `float()`, `str()`, `bool()`

```python
# ✅ Chuyển đổi thành công

# String → Integer
s = "123"
n = int(s)
print(f'int("123") = {n}, kiểu: {type(n)}')

# Integer → String
n = 456
s = str(n)
print(f'str(456) = "{s}", kiểu: {type(s)}')

# Integer → Float
n = 10
f = float(n)
print(f'float(10) = {f}')

# Float → Integer (CẮT phần thập phân, KHÔNG làm tròn)
f = 3.9
n = int(f)
print(f'int(3.9) = {n}  ← Cắt bỏ .9, KHÔNG làm tròn lên 4!')
```

```python
# ❌ LỖI: Chuyển string không phải số sang int
s = "hello"
n = int(s)  # ValueError: invalid literal for int()
```

```python
# ❌ LỖI: Chuyển string có chữ số lẫn chữ cái
s = "12abc"
n = int(s)  # ValueError
```

```python
# ❌ LỖI: Chuyển string float sang int trực tiếp
s = "3.14"
n = int(s)  # ValueError! Phải qua float trước
# Cách đúng: int(float(s))
```

```python
# ✅ Cách đúng: String float → int qua 2 bước
s = "3.14"
n = int(float(s))  # Bước 1: float → 3.14, Bước 2: int → 3
print(f'int(float("3.14")) = {n}')
```

```python
# ⚠️ bool() - Quy tắc: giá trị "rỗng" hoặc 0 → False, còn lại → True
print("--- Các giá trị False ---")
print(f'bool(0) = {bool(0)}')
print(f'bool(0.0) = {bool(0.0)}')
print(f'bool("") = {bool("")}')
print(f'bool([]) = {bool([])}')  # list rỗng
print(f'bool(None) = {bool(None)}')

print("\n--- Các giá trị True ---")
print(f'bool(1) = {bool(1)}')
print(f'bool(-1) = {bool(-1)}')
print(f'bool("abc") = {bool("abc")}')
print(f'bool(" ") = {bool(" ")}')  # Space cũng là True!
```

## 4. Nhập Dữ Liệu - `input()`

⚠️ **Quy tắc quan trọng**: `input()` **LUÔN trả về kiểu `str`**  
→ Phải chuyển đổi kiểu nếu cần số!

```python
# input() luôn trả về string
age = input("Nhập tuổi: ")
print(f"age = {age}")
print(f"Kiểu: {type(age)}")
# Dù nhập số, vẫn là string!
```

```python
# ❌ LỖI: Cộng input không chuyển kiểu
a = input("Nhập số a: ")  # Nhập "5"
b = input("Nhập số b: ")  # Nhập "3"
print(f"a + b = {a + b}")  # Kết quả: "53" (nối chuỗi) chứ KHÔNG phải 8!
```

```python
# ✅ Cách đúng: Chuyển sang int/float trước
a = int(input("Nhập số a: "))
b = int(input("Nhập số b: "))
print(f"a + b = {a + b}")  # Kết quả: 8 ✓
```

## 5. Xuất Dữ Liệu - `print()`

Các cách định dạng output:

```python
name = "Hải"
age = 20
gpa = 3.567

# Cách 1: Nối bằng dấu phẩy
print("Tên:", name, "- Tuổi:", age)

# Cách 2: f-string (khuyên dùng ✅)
print(f"Tên: {name} - Tuổi: {age}")

# Cách 3: format()
print("Tên: {} - Tuổi: {}".format(name, age))

# Cách 4: % operator (cũ)
print("Tên: %s - Tuổi: %d" % (name, age))
```

```python
# Định dạng số thập phân
gpa = 3.567

print(f"GPA: {gpa}")       # 3.567 (mặc định)
print(f"GPA: {gpa:.2f}")   # 3.57  (2 chữ số thập phân)
print(f"GPA: {gpa:.1f}")   # 3.6   (1 chữ số thập phân)
print(f"GPA: {gpa:.0f}")   # 4     (0 chữ số, làm tròn)
```

```python
# print() nâng cao: sep và end
print(1, 2, 3, sep="-")         # 1-2-3
print("Hello", end=" ")         # Không xuống dòng
print("World")                   # Hello World (cùng dòng)
```

## 6. Bài Tập Tổng Hợp

```python
# Bài tập: Tính diện tích và chu vi hình chữ nhật
width = 5
height = 3

area = width * height
perimeter = 2 * (width + height)

print(f"Hình chữ nhật {width}x{height}")
print(f"Diện tích: {area}")
print(f"Chu vi: {perimeter}")
```

```python
# Bài tập: Đổi nhiệt độ Celsius ↔ Fahrenheit
celsius = 37.5
fahrenheit = celsius * 9/5 + 32

print(f"{celsius}°C = {fahrenheit}°F")

# Đổi ngược lại
f_input = 100
c_output = (f_input - 32) * 5/9
print(f"{f_input}°F = {c_output:.2f}°C")
```
