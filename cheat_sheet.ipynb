## 2. Logic Cốt Lõi & Thuật Toán Cơ Bản (Core Logic)
Bao quát: Min/Max 3 số, Hoán vị, Chẵn/Lẻ, Năm nhuận, Đảo ngược số.

```python
# 1. Tìm Min/Max của 3 số (Dùng if-else)
def find_min_3(a, b, c):
    min_val = a
    if b < min_val: min_val = b
    if c < min_val: min_val = c
    return min_val

# 2. Hoán vị 2 biến (Swap)
a, b = 5, 10
a, b = b, a

# 3. Kiểm tra Chẵn/Lẻ (Odd/Even)
def is_even(n): return n % 2 == 0

# 4. Kiểm tra Năm nhuận (Leap Year)
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# 5. Đảo ngược một số nguyên (VD: 123 -> 321)
def reverse_integer(n):
    res = 0
    temp = abs(n)
    while temp > 0:
        res = res * 10 + temp % 10
        temp //= 10
    return res if n >= 0 else -res

# 6. Đếm số lượng ước số
def count_divisors(n):
    count = 0
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            if i*i == n: count += 1
            else: count += 2
    return count

# 7. Kiểm tra tính chia hết (Divisibility)
def is_divisible(a, b):
    """Kiểm tra a có chia hết cho b không"""
    if b == 0: return False
    return a % b == 0

# 8. Tính tổng các ước số (Sum of Divisors)
def sum_divisors(n):
    s = 0
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            if i*i == n: s += i
            else: s += i + (n // i)
    return s

# 9. Kiểm tra số chính phương (Perfect Square)
def is_perfect_square(n):
    if n < 0: return False
    root = int(math.sqrt(n))
    return root * root == n
```

## 3. Toán Học & Thuật Toán Số Học
Bao quát: Số nguyên tố, GCD, LCM, Giai thừa, Fibonacci, Số hoàn hảo.

```python
def is_prime(n):
    """Kiểm tra số nguyên tố"""
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True

def gcd(a, b):
    """ƯớC chung lớn nhất (Greatest Common Divisor)"""
    while b:
        a, b = b, a % b
    return a

def lcm(a, b):
    """Bội chung nhỏ nhất (Least Common Multiple)"""
    if a == 0 or b == 0: return 0
    return abs(a * b) // gcd(a, b)

def factorial(n):
    """Tính giai thừa n!"""
    if n < 0: return None
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res

def is_perfect_number(n):
    """Số hoàn hảo (tổng ước bằng chính nó, VD: 6, 28)"""
    if n < 1: return False
    total = 0
    for i in range(1, n):
        if n % i == 0: total += i
    return total == n

def fibonacci_list(n):
    """Trả về list n số Fibonacci đầu tiên"""
    if n <= 0: return []
    if n == 1: return [0]
    fib = [0, 1]
    while len(fib) < n:
        fib.append(fib[-1] + fib[-2])
    return fib
```

## 4. Xử Lý Chuỗi (String Mastery)
Bao quát: Đếm số, tổng số, đảo chuỗi, Palindrome, làm sạch chuỗi.

```python
def sum_digits_in_string(s):
    """Tổng các ký tự là chữ số trong chuỗi"""
    return sum(int(char) for char in s if char.isdigit())

def count_vowels(s):
    """Đếm số nguyên âm (u, e, o, a, i)"""
    vowels = "ueoaiUEOAI"
    return sum(1 for char in s if char in vowels)

def is_palindrome(s):
    """Kiểm tra chuỗi đối xứng (không phân biệt hoa thường, bỏ khoảng trắng)"""
    clean_s = "".join(s.split()).lower()
    return clean_s == clean_s[::-1]

def reverse_words(s):
    """Đảo ngược các từ trong chuỗi (VD: 'Hello World' -> 'World Hello')"""
    return " ".join(s.split()[::-1])

def remove_duplicate_chars(s):
    """Bỏ các ký tự trùng lặp, giữ lại thứ tự xuất hiện đầu tiên"""
    seen = set()
    result = []
    for char in s:
        if char not in seen:
            seen.add(char)
            result.append(char)
    return "".join(result)
```

## 5. Xử Lý Danh Sách & Thuật Toán Sắp Xếp (List Operations)
Bao quát: Sắp xếp, tìm kiếm, trung bình, min/max.

```python
# 1. Tìm Min/Max trong List (Thủ công)
def find_min_max_manual(lst):
    if not lst: return None, None
    min_v = max_v = lst[0]
    for x in lst:
        if x < min_v: min_v = x
        if x > max_v: max_v = x
    return min_v, max_v

# 2. Tìm kiếm tuyến tính (Linear Search)
def linear_search(lst, target):
    for i in range(len(lst)):
        if lst[i] == target: return i
    return -1

def bubble_sort(arr):
    """Sắp xếp nổi bọt (Bubble Sort)"""
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def selection_sort(arr):
    """Sắp xếp chọn (Selection Sort)"""
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def get_average(lst):
    """Tính trung bình cộng"""
    return sum(lst) / len(lst) if lst else 0

def find_second_largest(lst):
    """Tìm phần tử lớn thứ hai"""
    unique_sorted = sorted(list(set(lst)), reverse=True)
    return unique_sorted[1] if len(unique_sorted) > 1 else None

# 6. Đếm số lần xuất hiện (Thủ công)
def count_occurrences_manual(lst, target):
    count = 0
    for x in lst:
        if x == target: count += 1
    return count

# 7. Lọc phần tử (Thủ công - Filter)
def filter_even_numbers(lst):
    res = []
    for x in lst:
        if x % 2 == 0: res.append(x)
    return res

# 8. Kiểm tra List đã sắp xếp chưa
def is_sorted(lst):
    for i in range(len(lst) - 1):
        if lst[i] > lst[i+1]: return False
    return True
```

## 6. Từ Điển & Tập Hợp (Dict & Set) - CHI TIẾT
Bao quát: Đếm tần suất, sắp xếp dict, phép toán tập hợp.

```python
def frequency_count(lst):
    """Đếm tần suất xuất hiện của các phần tử"""
    freq = {}
    for item in lst:
        freq[item] = freq.get(item, 0) + 1
    return freq

def get_most_frequent(lst):
    """Tìm phần tử xuất hiện nhiều nhất"""
    freq = frequency_count(lst)
    if not freq: return None
    return max(freq, key=freq.get)

def sort_dict_by_value(d, reverse=True):
    """Sắp xếp dictionary theo giá trị"""
    return dict(sorted(d.items(), key=lambda item: item[1], reverse=reverse))

def merge_dictionaries(d1, d2):
    """Gộp 2 dictionary (Python 3.9+ có d1 | d2)"""
    res = d1.copy()
    res.update(d2)
    return res

# --- SET OPERATIONS (TẬP HỢP) ---
def set_tricks(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return {
        "union": s1 | s2,          # Hợp
        "intersection": s1 & s2,   # Giao (phần tử chung)
        "difference": s1 - s2,     # Hiệu (có ở s1 nhưng không có ở s2)
        "symmetric_diff": s1 ^ s2  # Những phần tử chỉ thuộc 1 trong 2 tập
    }
```

## 7. Xử Lý File (File I/O) - CHI TIẾT
Bao quát: Đọc CSV, xử lý đường dẫn, ghi đè/ghi thêm.

```python
import os

def file_exists(filename):
    return os.path.exists(filename)

def read_csv_manual(filename, sep=',', skip_header=False):
    """Đọc file CSV thủ công (không dùng thư viện pandas/csv)"""
    data = []
    if not os.path.exists(filename): return []
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        if skip_header: lines = lines[1:]
        for line in lines:
            line = line.strip()
            if line:
                row = [item.strip() for item in line.split(sep)]
                data.append(row)
    return data

def save_data_to_file(filename, data, mode='w'):
    """Lưu list lồng nhau hoặc list đơn vào file"""
    with open(filename, mode, encoding='utf-8') as f:
        for row in data:
            if isinstance(row, list):
                line = ", ".join(map(str, row))
            else:
                line = str(row)
            f.write(line + '\n')

def read_file_as_string(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        return f.read()
```

## 8. Lập Trình Hướng Đối Tượng (OOP Template) - CHI TIẾT
Bao quát: Class, Encapsulation, Inheritance, List of Objects handling.

```python
class Student:
    def __init__(self, s_id, name, gpa):
        self.__id = s_id    # Thuộc tính private
        self.name = name
        self.gpa = float(gpa)
    
    # Getter cho private attribute
    def get_id(self):
        return self.__id
    
    def __str__(self):
        return f"{self.__id} | {self.name} | {self.gpa:.2f}"

# --- QUẢN LÝ DANH SÁCH ĐỐI TƯỢNG (DÀNH CHO PE) ---

def find_by_id(obj_list, search_id):
    """Tìm đối tượng trong list theo ID"""
    for obj in obj_list:
        if obj.get_id() == search_id:
            return obj
    return None

def sort_by_gpa(obj_list, descending=True):
    """Sắp xếp list đối tượng theo GPA"""
    return sorted(obj_list, key=lambda x: x.gpa, reverse=descending)

def filter_by_gpa_range(obj_list, min_v, max_v):
    """Lọc đối tượng trong khoảng GPA"""
    return [x for x in obj_list if min_v <= x.gpa <= max_v]

# --- Kế thừa (Inheritance) ---
class GradStudent(Student):
    def __init__(self, s_id, name, gpa, major):
        super().__init__(s_id, name, gpa)
        self.major = major
        
    def __str__(self):
        # Gọi __str__ của lớp cha và thêm thuộc tính mới
        return super().__str__() + f" | Major: {self.major}"
```
